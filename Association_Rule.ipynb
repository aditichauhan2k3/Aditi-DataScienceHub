Association Rule Mining:

Implement an Apriori algorithm using tool like python with libraries such as Pandas and Mlxtend etc.

 Apply association rule mining techniques to the pre-processed dataset to discover interesting relationships between products purchased together.

Set appropriate threshold for support, confidence and lift to extract meaning full rules.

```python
# 2. Convert each transaction row into a list of items
transactions = df.iloc[:, 0].apply(
    lambda x: [item.strip() for item in str(x).split(',') if item.strip()]
)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

#Encode the transaction
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

# Apply Apriori
frequent_itemsets = apriori(df,min_support=0.02, use_colnames=True)

# Generate rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

#sort and view top rules
rules = rules.sort_values(by='lift', ascending=False)
print(rules.head())
```

INTERVIEW QUESTIONS & ANSWERS

1. What is Lift and why is it important in Association Rules?

Lift measures the strength of an association between two items by comparing their co-occurrence with what would be expected if they were independent.

FORMULA:  Lift(A→B)=Confidence(A→B) / Support(B)​


Why it is important:

Lift > 1 → Positive association (items are bought together more than expected)

Lift = 1 → No association

Lift < 1 → Negative association

Helps identify meaningful and useful rules, not just frequent ones

2. What is Support and Confidence? How do you calculate them?

**Support**

Support indicates how frequently an itemset appears in the dataset.

Support(A)= Number of transactions containing A​ / Total number of transactions

Measures popularity of an itemset



**Confidence**

Confidence measures how often item B is purchased when item A is purchased.

Confidence(A→B)= Support(A∩B) / Support(A)

Measures reliability of the rule​


3. What are some limitations or challenges of Association Rule Mining?


Generates a very large number of rules

Computationally expensive for large datasets

Choosing appropriate support and confidence thresholds is difficult

Does not imply causation, only association

Rare but important items may be ignored

Interpretation requires domain knowledge
